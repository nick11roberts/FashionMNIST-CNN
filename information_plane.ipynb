```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1 = np.fromfile('iXT1.dat', dtype=float)
t2 = np.fromfile('iXT2.dat', dtype=float)
t3 = np.fromfile('iXT3.dat', dtype=float)
t4 = np.fromfile('iXT4.dat', dtype=float)

yt1 = np.fromfile('iTY1.dat', dtype=float)
yt2 = np.fromfile('iTY2.dat', dtype=float)
yt3 = np.fromfile('iTY3.dat', dtype=float)
yt4 = np.fromfile('iTY4.dat', dtype=float)
```

```python
len(t1)
```

```python
f = plt.figure(figsize=(12, 12))
axes = f.add_subplot(111)
axes = np.array([[axes]])

cmap = plt.get_cmap('gnuplot')
colors = [cmap(i) for i in np.linspace(0, 1, len(t1))]

#cbaxes = f.add_axes([1.0, 0.1, 0.03, 0.8])  # This is the position for the colorbar
#cb = plt.colorbar(axes[0, 0], cax=cbaxes)

for i in range(0, len(t1)):
    axes[0, 0].plot(([t4[i], t3[i], t2[i], t1[i]]), 
                    ([yt4[i], yt3[i], yt2[i], yt1[i]]), 
                    marker='o', linestyle='-', markersize=12, 
                    markeredgewidth=0.01, linewidth=0.2, 
                    color=colors[i])

plt.title('Information Plane')
plt.xlabel('I(X;Ti)')
plt.ylabel('I(Ti;Y)')
plt.show()

# Here, warmer colors represent more epochs
```
