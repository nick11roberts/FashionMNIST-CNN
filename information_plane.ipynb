```python
f = plt.figure(figsize=(12, 8))
axes = f.add_subplot(111)
axes = np.array([[axes]])

cmap = plt.get_cmap('gnuplot')
#For each epoch we have diffrenet color
colors = [cmap(i) for i in np.linspace(0, 1, len(t1))]

for i in range(0, len(t1)):
    axes[0, 0].plot([t4[i], t3[i], t2[i], t1[i]], 
                    [np.random.rand(1), np.random.rand(1)+10, np.random.rand(1)+20, np.random.rand(1)+30], 
                    marker='o', linestyle='-', markersize=12, 
                    markeredgewidth=0.01, linewidth=0.2, 
                    color=colors[i])
plt.show()
```
